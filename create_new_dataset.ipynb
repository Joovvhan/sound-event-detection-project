```python
import os
import glob
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
wav_file_paths = glob.glob('new_data/*.wav')
wav_file_paths.sort()
```

```python
new_dataset_dir = [path_name.replace('.wav', '') for path_name in wav_file_paths]
```

```python
[os.makedirs(dir_name, exist_ok=True) for dir_name in new_dataset_dir]
```

```python
os.listdir('new_data/')
```

```python
MAXLOOP = 10000
SR = 44100

for file_path in wav_file_paths:
    
    save_dir = file_path.replace('.wav', '')
    
    file_format = file_path.split('/')[-1].replace('.wav', '_{:05d}.wav')
    
    fs = librosa.get_samplerate(file_path)
    block_length = 5
    frame_length = fs * 1
    hop_length = frame_length
    
    audio_stream = np.array([])
    
    stream = librosa.core.stream(file_path, block_length, frame_length, hop_length)
    
    for i in range(MAXLOOP):
        
        random_second = np.random.uniform(1, 5)
        random_length_in_samples = int(np.floor(fs * random_second))
        
        try:
            if len(audio_stream) < random_length_in_samples:
                audio_stream = np.append(audio_stream, next(stream))
        except StopIteration:
            print('[END of WAV]')
            break
            
        buffer_to_save = audio_stream[:random_length_in_samples]
        audio_stream = audio_stream[random_length_in_samples:]
        
        if fs != SR:
            buffer_to_save = librosa.core.resample(buffer_to_save, fs, SR)
        
        save_file_name = os.path.join(save_dir, file_format.format(i))
        librosa.output.write_wav(save_file_name, buffer_to_save.astype(np.float32), SR)
        print('{}: [{}]'.format(save_file_name, buffer_to_save.shape))
        
        
        
```
